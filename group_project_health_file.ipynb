```python
#create new dataframe with 
#  
#select dataframe that only has region code, name, year and value but just for current health exp of gdp

##  create dataframe of only 'Domestic general government health expenditure (% of total government expenditure)') in series

health_domgenper_df = health_df.loc[(health_df['Series'] == 'Domestic general government health expenditure (% of total government expenditure)')]

health_domgenper_df.head()
```

```python
## export dataframe to excel file
#health_chper_df
#health_domgenper_df

#save dataframe with 
#Current health expenditure (% of GDP)'

writer = pd.ExcelWriter('health_chper.xlsx', engine='xlsxwriter')
health_chper_df.to_excel(writer, sheet_name='health_chper')
writer.save()
```
